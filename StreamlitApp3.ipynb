```python
%%writefile app.py
import streamlit as st
import tensorflow as tf
import numpy as np

#Tensorflow model prediction
def model_prediction(test_image):
  model=tf.keras.models.load_model('/content/drive/MyDrive/Colab Notebooks/Summer Training2024/Kaggle/Model5_model.h5')
  image=tf.keras.preprocessing.image.load_img(test_image,target_size=(180,180))
  input_arr=tf.keras.preprocessing.image.img_to_array(image)
  input_arr=np.array([input_arr])#Convert single image to batch
  predictions=model.predict(input_arr)
  return np.argmax(predictions)#REturn index of the max element

def fetch_calories(prediction):
  calorie={'Apple': 52,'Banana': 89,"Grapes":67,"Kiwi":61,"Mango":60,"Orange":47,"Pear":57,"Pineapple":50,"Pomegranate":68,"Watermelon":30,"Beetroot":43,"Bell Pepper":20,"Cabbage":25,"Capsicum":40,"Carrot":41,"Cauliflower":25,"Chilli Pepper":40,
           "Corn":86,"Cucumber":16,"Eggplant":25,"Garlic":111,"Ginger":80,"Jalepeno":28,"Lemon":29,"Lettuce":15,"Sweetcorn":86,"Onion":40,"Paprika":282,"Peas":81,"Potato":87,"Radish" : 16,"Soy Beans":172,"Spinach":23,"Sweetpotato":86,"Tomato":19,"Turnip":28}
  if prediction in calorie:
    return calorie.get(prediction)
  else:
    st.error("Sorry ! Calories not found")


#sidebar
st.sidebar.title("DashBoard")
app_mode=st.sidebar.selectbox("Select page",["Home","About Project","Prediction"])

#Main Page
if(app_mode=="Home"):
  st.header("Fruits and Vegetables Recognition Model")
  image_path="/content/drive/MyDrive/Colab Notebooks/Summer Training2024/Kaggle/Appimg.jpeg"
  st.image(image_path,width=400)

#About Project
elif(app_mode=='About Project'):
  st.header("About Project")
  st.subheader("About dataset")
  st.text("This dataset contains images of the following food items : ")
  st.code("Fruits - Banana, Apple, Pear, Grapes, Orange, Kiwi, Watermelon, Pomegranate, Pineapple, Mango")
  st.code("Vegetables -  Cucumber, Carrot, Capsicum, Onion, Potato, Lemon, Tomato, Radish, Beetroot, Cabbage, Lettuce, Spinach, Soybean, Cauliflower, Bell Pepper, Chilli Pepper, Turnip, Corn, Sweetcorn, Sweet Potato, Paprika, JalapeÃ±o, Ginger, Garlic, Peas, Eggplant")
  st.subheader("Content")
  st.text("This dataset contains three folders : ")
  st.text("1. train (100 images each)")
  st.text("2. test (10 images each)")
  st.text("3. validation (10 images each)")

#Prediction page
elif(app_mode=="Prediction"):
  st.header("Model Prediction")
  test_image=st.file_uploader("Choose an image : ")
  if(st.button("Show Image")):
    st.image(test_image,width=4,use_column_width=True)

  if(st.button("Predict")):
    fruits=['Apple','Banana','Grapes','Kiwi','Mango','Orange','Pear','Pineapple','Pomegranate','Watermelon']
    vegetables=['Beetroot','Bell Pepper','Cabbage','Capsicum','Carrot','Cauliflower','Chilli Pepper','Corn','Cucumber','Eggplant','Garlic','Ginger','Jalepeno','Lemon','Lettuce','Sweetcorn','Onion','Paprika','Peas','Potato','Raddish','Soy Beans','Spinach','Sweetpotato','Tomato','Turnip']
    st.write("Our prediction")
    result_index=model_prediction(test_image)
    labels=['Apple','Banana','Beetroot','Bell pepper','Cabbage','Capsicum','Carrot','Cauliflower','Chilli pepper','Corn','Cucumber','Eggplant','Garlic','Ginger','Grapes','Jalepeno','Kiwi','Lemon','Lettuce','Mango','Onion','Orange','Paprika','Pear','Peas','Pineapple','Pomegranate','Potato','Raddish','Soy beans','Spinach','Sweetcorn','Sweetpotato','Tomato','Turnip','Watermelon']
    prediction = labels[result_index]
    if prediction in vegetables:
      st.info('Category   :   Vegetable')
    else:
      st.info("Category   :   Fruit")
    st.success("Model Prediction : {}".format(prediction))
    cal=fetch_calories(prediction)
    if cal:
        st.warning("Calories in every 100gms  :  "+str(cal)+"cal")
```
